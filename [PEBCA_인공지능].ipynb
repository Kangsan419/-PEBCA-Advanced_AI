```python
from tensorflow import keras 
import numpy as np
```

```python
x =np.array([0,1,2,3,4,5],dtype='float64')
y=x*2+1
print(y)
```

```python
model = keras.models.Sequential()
#1차원 데이터 아웃풋, inputshape이 1개만 들어온다 
model.add(keras.layers.Dense(1,input_shape=(1,)))
model.compile('Adadelta','mse')
```

```python
model.fit(x,y,epochs=1000)
```

```python
model.predict([7,10,99]) #with 'sgd' 100번 
```

```python
model.predict([7,10,99]) #with 'sgd' 1000번 
```
